```python
for df in (train , test): 
    df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
    df.sort_values('timestamp', inplace=True)
    df.reset_index(drop=True, inplace=True)
```

```python
print("rows -> train:", len(train), "| test:", len(test))
print("nulls(train):", train.isnull().sum().to_dict())
print("nulls(test):",  test.isnull().sum().to_dict())
print("time monotonic ->", train['timestamp'].is_monotonic_increasing, test['timestamp'].is_monotonic_increasing)
```

```python
dt = train['timestamp'].diff().dropna().dt.total_seconds()
fs = 1/dt[dt>0].median()
print("median delta time (s):", dt[dt>0].median(), " ~ fs (Hz):", fs)
```
