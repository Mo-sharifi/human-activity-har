```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
```

```python
print("rows -> train:", len(train), "| test:", len(test))
print("nulls(train):", train.isnull().sum().to_dict())
print("nulls(test):",  test.isnull().sum().to_dict())
print("time monotonic ->", train['timestamp'].is_monotonic_increasing, test['timestamp'].is_monotonic_increasing)
```

```python
dt = train['timestamp'].diff().dropna().dt.total_seconds()
fs = 1/dt[dt>0].median()
print("median delta time (s):", dt[dt>0].median(), " ~ fs (Hz):", fs)
```

```python
if 'label' in train.columns:
    print("label counts:\n", train['label'].value_counts().sort_index())
```

```python
WINDOW = 200   
STRIDE = 50    
MIN_PURITY = 0.7  
win_labels = []
win_idx = []   

n = len(train)
for s in range(0, n - WINDOW + 1, STRIDE):
    e = s + WINDOW
    w_lbl = train['label'].iloc[s:e]
    vc = w_lbl.value_counts(normalize=True)
    maj, purity = vc.idxmax(), vc.max()
    if purity >= MIN_PURITY:
        win_labels.append(maj)
        win_idx.append((s, e))

win_labels = np.array(win_labels)
win_idx = np.array(win_idx)

print("number of windows", len(win_idx))
print("label distributed:")
print(pd.Series(win_labels).value_counts().sort_index())
```

```python
SENS = ['back_x','back_y','back_z','thigh_x','thigh_y','thigh_z']

def window_features(df, s, e):
    A = df[SENS].iloc[s:e].values  
    feats = np.hstack([
        A.mean(0),               
        A.std(0),               
        A.min(0),               
        A.max(0),                
        (A**2).mean(0),         
    ])
    return feats  


X_feats = np.vstack([window_features(train, s, e) for (s,e) in win_idx])
y_win   = win_labels.astype(int)

print("X_feats.shape:", X_feats.shape)
print("y_win.shape:", y_win.shape)
```

```python
cut = int(0.8 * len(X_feats))
X_train, X_val = X_feats[:cut], X_feats[cut:]
y_train, y_val = y_win[:cut], y_win[cut:]

print("Train:", X_train.shape, " Val:", X_val.shape)
```

```python
clf = RandomForestClassifier(
    n_estimators=200, 
    class_weight='balanced', 
    random_state=42, 
    n_jobs=-1
)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_val)
f1 = f1_score(y_val, y_pred, average='macro')
print("Validation macro-F1:", round(f1, 3))
```

```python
WINDOW, STRIDE = 200, 50
SENS = ['back_x','back_y','back_z','thigh_x','thigh_y','thigh_z']

def window_features(df, s, e):
    A = df[SENS].iloc[s:e].values
    return np.hstack([A.mean(0), A.std(0), A.min(0), A.max(0), (A**2).mean(0)])

Xt, idx_te = [], []
n = len(test)
for s in range(0, n - WINDOW + 1, STRIDE):
    e = s + WINDOW
    Xt.append(window_features(test, s, e))
    idx_te.append((s, e))

Xt = np.vstack(Xt)           
idx_te = np.array(idx_te)    
Xt.shape, idx_te.shape
```

```python
win_pred = clf.predict(Xt)   
classes = np.sort(np.unique(y_win))      
class2col = {c:i for i,c in enumerate(classes)}
```

```python
K = len(classes)
votes = np.zeros((len(test), K), dtype=np.int32)

for (s, e), lbl in zip(idx_te, win_pred):
    votes[s:e, class2col[lbl]] += 1

row_pred_col = votes.argmax(axis=1)
row_pred_lbl = classes[row_pred_col]

test['label'] = row_pred_lbl.astype(int)
test.head()
```
